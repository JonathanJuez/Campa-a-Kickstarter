# Escalado y normalizacion

```python
from scipy import stats

# for min_max scaling
from sklearn.preprocessing import minmax_scale
#sklearn.preprocessing.minmax_scale

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

#conforme a lo analizado es las graficas de dispersion se debe escalar la variable goal USD ya que esta variable contiene valores en varias denominaciones de la moneda en cada pais 
# select the usd_goal_real column
usd_goal = dfle.goal_usd

# scale the goals from 0 to 1
scaled_goal = minmax_scale(usd_goal)#

#retornar valores al dataset
dfle.goal_usd = scaled_goal

# plot the original & scaled data together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(dfle.goal_usd, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_goal, ax=ax[1])
ax[1].set_title("Scaled data")
```

```python
usd_recaudado = dfle.usd_pledged

# scale the goals from 0 to 1
scaled_recaudado = minmax_scale(usd_recaudado)#, columns = [0])

#retornar valores al dataset 
dfle.usd_pledged = scaled_recaudado

# plot the original & scaled data together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(dfle.usd_pledged, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_recaudado, ax=ax[1])
ax[1].set_title("Scaled data")
```
